```python
def treat_outliers(df, method='iqr', threshold=1.5):

    numeric_columns = df.select_dtypes(include=['int', 'float']).columns

    for col in numeric_columns:
        if method == 'iqr':
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - threshold * IQR
            upper_bound = Q3 + threshold * IQR
            df[col] = df[col].apply(lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x)

    return df


df1 = treat_outliers(df)
```

```python
int_columns = df1.select_dtypes(include=['int','float']).columns

# Plot boxplots for each integer column
for column in int_columns:
    plt.figure(figsize=(6,4))
    df.boxplot(column=column)
    plt.title(f"Boxplot of {column}")
    plt.ylabel("Values")
    plt.show()
```

**Lets do the scaling of various features**

```python
from sklearn.preprocessing import MinMaxScaler

numeric_columns = df_mobile.select_dtypes(include=['int', 'float']).columns

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Apply Min-Max Scaling to selected columns
df_mobile[numeric_columns] = scaler.fit_transform(df_mobile[numeric_columns])
```

```python
#We do not need the preferred device column as we have only mobile

df_mobile.drop(['preferred_device'], axis = 1, inplace = True)
df_mobile.drop(['UserID'], axis = 1, inplace = True)
```

```python
#Label encoding mobile dataset

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df_mobile['preferred_location_type'] = label_encoder.fit_transform(df_mobile['preferred_location_type'])
df_mobile['Taken_product'] = label_encoder.fit_transform(df_mobile['Taken_product'])
df_mobile['working_flag'] = label_encoder.fit_transform(df_mobile['working_flag'])

df_mobile['Taken_product'].value_counts()
```

```python
df_mobile_target = df_mobile['Taken_product']
df_mobile_train = df_mobile.drop(['Taken_product'], axis = 1)
```

```python
df_mobile_train.head()
```

```python
#train test split

from sklearn.model_selection import train_test_split
(X_train_mob, X_test_mob, Y_train_mob, Y_test_mob) = train_test_split(df_mobile_train,df_mobile_target, train_size = 0.8)
```

```python
from sklearn.linear_model import LogisticRegression
logisticRegr = LogisticRegression()
logisticRegr.fit(X_train_mob, Y_train_mob)
score = logisticRegr.score(X_test_mob, Y_test_mob)
print(score)
LR_mob_pred = logisticRegr.predict(X_test_mob)
LR_mob_pred
```

```python
from sklearn.metrics import confusion_matrix
LR_cm = confusion_matrix(Y_test_mob, LR_mob_pred)
print(LR_cm)

plt.figure(figsize=(6,6))
sns.heatmap(LR_cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(score)
plt.title(all_sample_title, size = 15);
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 1000, random_state = 42, oob_score = True)
rf.fit(X_train_mob, Y_train_mob)

from sklearn.metrics import accuracy_score

RF_mob_pred = rf.predict(X_test_mob).round()
#rf.oob_score_
accuracy_score(Y_test_mob, RF_mob_pred)
score = rf.score(X_test_mob, Y_test_mob)
score

```
